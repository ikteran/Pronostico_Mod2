```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Modelo': ['MARS', 'LSTM', 'Random Forest', 'SSA'],
    'RMSE': [4.25, 3.79, 5.81, 5.21],
    'MAE': [3.68, 1.89, 4.50, 4.24],
}

# Crear DataFrame
df_metrics = pd.DataFrame(data)
df_metrics.set_index('Modelo', inplace=True)

# Gráfico
ax = df_metrics.plot(kind='bar', figsize=(10, 6))
plt.title('Comparación de RMSE y MSE: Predicciones en el activo META')
plt.ylabel('Valor de la métrica')
plt.xlabel('Modelo')
plt.xticks(rotation=0)
plt.legend(title='Métrica')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
